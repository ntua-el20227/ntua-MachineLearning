### 3. Προεπεξεργασία συνόλου δεδομένων
💡`Χρησιμοποιήστε το Column Transformer για τη δημιουργία και την εφαρμογή χωριστών μετασχηματιστών για αριθμητικά και κατηγορικά δεδομένα.`




```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest,chi2
```

3.α.  Θα χρησιμοποιήσετε όλα τα χαρακτηριστικά του συνόλου δεδομένων για την εκπαίδευση των ταξινομητών ή θα επιλέξετε κάποια από αυτά;  Θα κάνετε κάποια συνένωση κάποιων χαρακτηριστικών για δημιουργία νέων χαρακτηριστικών για το μοντέλο σας;


_We tried to remove the id column because the id of samples is irrelevant to the prediction_

```python
ids = trainSet.pop('id') # remove id column for the  trainSet
```

_We will also create a new column called Season which will replace the Date column as it is more important to extract the season we have as data to predict the RainTommorrow rather than the exact day and month which might train our model with not so important details_

```python
# This is the function we create for extracting the season from each date
def get_season(date):
    if date.month in [3, 4, 5]:
        return 'Spring'
    elif date.month in [6, 7, 8]:
        return 'Summer'
    elif date.month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'
```

```python

# Convert the 'Date' column to datetime
trainSet['Date'] = pd.to_datetime(trainSet['Date'])
# Extract the season from the 'Date' column
trainSet['Season'] = trainSet['Date'].apply(get_season)
trainSet.drop(['Date'], axis=1, inplace=True) # drop the 'Date' column
```

```python
#We do the same for the test set
testSet['Date'] = pd.to_datetime(testSet['Date'])
testSet['Season'] = testSet['Date'].apply(get_season)
testSet.drop(['Date'], axis=1, inplace=True) # drop the 'Date' column
```

3.β. Υπάρχουν απουσιάζουσες τιμές; Γράψτε κατάλληλο κώδικα ώστε να χειριστείτε αυτές τις τιμές.

```python
# Show missing values
trainSet.isnull().sum()
```

```python
#Imputers for missing values
from sklearn.impute import KNNImputer

step1_num = ('imputer', SimpleImputer(strategy='mean')) # handle missing values of numeric features
step1_cat = ('imputer', SimpleImputer(strategy='most_frequent')) # handle missing values of categorical features
```

3.γ. Γράψτε κώδικα για την κατάλληλη μετατροπή των κατηγορικών μεταβλητών ώστε να μπορούν να τους διαχειριστούν οι ταξινομητές που θα χρησιμοποιήσετε.

```python
step2_cat = ('scaler', OneHotEncoder(handle_unknown='ignore')) # handle categorical features
```

3.δ. Γράψτε κώδικα για την κλιμάκωση των χαρακτηριστικών, αν την θεωρείτε απαραίτητη

```python
step2_num = ('scaler', StandardScaler()) # handle numeric features
```

3.ε. Εκτελέστε όλα τα προηγούμενα βήμα προεκπαίδευσης (χρήση μετασχηματιστών) ώστε να είναι "καθαρό" το αρχικό σύνολο για να χρησιμοποιηθεί για την εκπαίδευση των ταξινομητών.

```python
# Create a transformer
numeric_transformer = Pipeline(steps=[step1_num,step2_num])
categorical_transformer = Pipeline(steps=[step1_cat,step2_cat])
```

3.στ. Αφού έχετε "καθαρίσει" το αρχικό σύνολο εκπαίδευσης, γράψτε τον κατάλληλο κώδικα ώστε από το αρχικό σύνολο εκπαίδευσης να δημιουργήσετε τα σύνολα Χ και y.

💡 `Η στήλη 'RainTomorrow' είναι η τιμή που θέλουμε να προβλέψει ο ταξινομητής μας.`



```python
y = trainSet.pop('RainTomorrow')
X = trainSet
```

```python
X_test = testSet # no RainTomorrow label in the test set
```

```python

numeric_features = ['MinTemp', 
                    'Evaporation', 'Sunshine','WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm',
                    'Humidity3pm','Pressure9am', 'Pressure3pm', 'Temp3pm', 'Cloud3pm','RainToday','Cloud9am','Humidity9am','MaxTemp','Temp9am']
categorical_features = ['Season','Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm']
```

```python
print(numeric_features)
print(categorical_features)
```

```python
preprocessor = ColumnTransformer( transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)] )
```

3.ζ. Διαχωρίστε το σύνολο δεδομένων σε σύνολο εκπαίδευσης (train set) και σε σύνολο επικύρωσης (validation set) (ο διαχωρισμός να είναι train set 70% και validation set 30%).


```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

preproc_X_train = preprocessor.fit_transform(X_train)
```

### 4. Εκπαίδευση μοντέλου με default τιμές στις παραμέτρους των ταξινομητών

Θα εκπαιδεύσουμε τους ακόλουθους ταξινομητές με το σύνολο εκπαίδευσης των δεδομένων μας:
1. Naive Bayes
2. KNeighborsClassifier
3. LogisticRegression
4. MLP με ένα κρυφό επίπεδο
5. SVC
6. Decision Tree
7. Random Forest


```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

 4.α. Εκπαιδεύστε (fit)  και τους  7 ταξινομητές που προαναφέρθηκαν (εμφανίστε τα δείγματα από το train set μαζί με την ετικέτα τους στην είσοδο του κάθε ταξινομητή), χρησιμοποιώντας τις default τιμές για όλες τις παραμέτρους τους.

```python
classifiers = {
    'GaussianNaiveBayes': GaussianNB(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'LogisticRegression': LogisticRegression(),
    'MLPClassifier': MLPClassifier(),
    'SVC': SVC(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'RandomForestClassifier': RandomForestClassifier()
} # dictionary of classifiers with their default parameters

for name, clf in classifiers.items():
    clf.fit(preproc_X_train.toarray(), y_train)
    print("Fitted", name)
```

4.β. Εφαρμόστε στα εκπαιδευμένα μοντέλα από το validation set μόνο τα δείγματα χωρίς την ετικέτα τους (predict).


```python
preproc_X_val = preprocessor.fit_transform(X_val)

predictions_val = {}
for name, clf in classifiers.items():
    print("Predicting", name)
    predictions_val[name] = clf.predict(preproc_X_val.toarray())
```


4.γ. Συγκρίνετε την έξοδο του κάθε μοντέλου σε σχέση με τις αντίστοιχες ετικέτες του validation set και αξιολογήστε την επιδοσή τους χρησιμοποιώντας το F1 score.


```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
```

```python
# compute the classification report
for name, clf in classifiers.items():
    print("Classification report for", name)
    print(classification_report(y_val, predictions_val[name]))
```

4.δ.  Αξιολογήστε συνολικά την επίδοση των μοντέλων χρησιμοποιώντας κάποιο γράφημα (π.χ. ιστόγραμμα, bar plot) και σχολιάστε ποιο μοντέλο είχε την καλύτερη επίδοση.


```python
# Calculate accuracy for each model
accuracy_scores = []
maximum_accuracy = 0
for name, val in predictions_val.items():
    accuracy_scores.append(accuracy_score(y_val, val))
    maximum_accuracy = max(maximum_accuracy, accuracy_score(y_val, val))
    if (maximum_accuracy == accuracy_score(y_val, val)):
        best_model_name_onAccuracy = name
        best_model_onAccuracy = classifiers[name]

# Plot accuracy for each model
plt.bar(classifiers.keys(), accuracy_scores)
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Performance')
plt.xticks(rotation=90)
plt.show()
```

```python

# Calculate precision for each model
precision_scores = []
maximum_precision = 0
for name, val in predictions_val.items():
    precision_scores.append(precision_score(y_val, val))
    maximum_precision = max(maximum_precision, precision_score(y_val, val))
    if (maximum_precision == precision_score(y_val, val)):
        best_model_name_onPrecision = name
        best_model_onPrecision = classifiers[name]

# Plot accuracy for each model
plt.bar(classifiers.keys(), precision_scores)
plt.xlabel('Models')
plt.ylabel('Precision')
plt.title('Model Performance')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate f1_score for each model
from sklearn.metrics import f1_score
f1_scores = []
maximum_f1_score = 0
best_model_name_onF1 = ""
for name, val in predictions_val.items():
    f1_scores.append(f1_score(y_val, val))
    maximum_f1_score = max(maximum_f1_score, f1_score(y_val, val))
    if (maximum_f1_score == f1_score(y_val, val)):
        best_model_name_onF1 = name
        best_model_onF1 = classifiers[name]

# Plot f1_score for each model
plt.bar(classifiers.keys(), f1_scores)
plt.xlabel('Models')
plt.ylabel('f1_score')
plt.title('Model Performance')
plt.xticks(rotation=90)
plt.show()
```

```python
print("Best model on Accuracy: ", best_model_name_onAccuracy);
print("Best model on Precision: ", best_model_name_onPrecision);
print("Best model on F1: ", best_model_name_onF1);
```

_After the first try I observed that the best estimator for MLPC was for a hidden layer of size=10. This could only mean that we overfitting our model on bigger sizes like 100 and 50 so for my next try I used the above hidden layers sizes_ 

```python
from sklearn.model_selection import GridSearchCV
# Create a dictionary to store the results
best_model = {}
best_params = {}
results = {}
grid = {}
best_estimator = {}
```
